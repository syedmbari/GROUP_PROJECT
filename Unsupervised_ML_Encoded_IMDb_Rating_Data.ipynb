```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
import pandas as pd
import tensorflow as tf
```

```python
# Generate our categorical variable list
Clean_imdb_data_df1_cat = Clean_imdb_data_df1.dtypes[Clean_imdb_data_df1.dtypes == "object"].index.tolist()


# Check the number of unique values in each column
Clean_imdb_data_df1[Clean_imdb_data_df1_cat].nunique()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
Clean_imdb_data_df1_encode_df = pd.DataFrame(enc.fit_transform(Clean_imdb_data_df1[Clean_imdb_data_df1_cat]))

# Add the encoded variable names to the dataframe
Clean_imdb_data_df1_encode_df.columns = enc.get_feature_names(Clean_imdb_data_df1_cat)
Clean_imdb_data_df1_encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
Clean_imdb_data_df1 = Clean_imdb_data_df1.merge(Clean_imdb_data_df1_encode_df,left_index=True, right_index=True)
Clean_imdb_data_df1 = Clean_imdb_data_df1.drop(Clean_imdb_data_df1_cat,1)
Clean_imdb_data_df1.head()
```

```python
# Remove Language_Status_as_English  target from features data
y = Clean_imdb_data_df1.Language_Status_as_English_yes.values
X = Clean_imdb_data_df1.drop(columns=["Language_Status_as_English_no","Language_Status_as_English_yes"]).values

# Split training/test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create the SVM model
svm = SVC(kernel='linear')
```

```python
# Train the model
svm.fit(X_train, y_train)
```

```python
# Evaluate the model of  Language_Status_as_English for viewers
y_pred = svm.predict(X_test_scaled)
print(f" SVM model accuracy: {accuracy_score(y_test,y_pred):.3f}")
```

```python
# Define the model - deep neural net
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 =  10
hidden_nodes_layer2 = 5

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))


# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Compile the Sequential model together and customize metrics
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Evaluate the model of  Language_Status_as_English for viewers
# Train the model 
fit_model = nn.fit(X_train_scaled, y_train, epochs=50) 
# Evaluate the model using the test data 
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
