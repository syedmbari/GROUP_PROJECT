```python
#iMPORTING DEPENDENCIES
import pandas as pd
import csv
import math
from sklearn.preprocessing import LabelEncoder
```

```python
#file_to_load iMDB_MOVIES
data_to_load = "GENRE_ENCODED.xlsx"
encoded_genre_df = pd.read_excel(data_to_load)
encoded_genre_df.head()
```

```python
#file_to_load iMDB_Data
data_to_load_imdb = "IMDb_data.csv"
raw_imdb_df = pd.read_csv(data_to_load_imdb)
raw_imdb_df.head()
```

```python
#Dropping genre
imdb_df = raw_imdb_df.drop(["genre 1", "genre 2", "genre 3"], axis = 1)
imdb_df.head()
```

```python
#encoding country and language
le = LabelEncoder()
imdb_df_c_l = imdb_df.copy()
imdb_df_c_l['country'] = le.fit_transform(imdb_df_c_l['country'])
imdb_df_c_l['language 1'] = le.fit_transform(imdb_df_c_l['language 1'])
imdb_df_c_l
```

```python
#Merging encoded_genre_df and imdb_df_c_l
encoded_imdb_df = pd.merge(imdb_df_c_l, encoded_genre_df, on="imdb_title_id", how="outer")

#Dropping imdb_title_id
encoded_imdb_df = encoded_imdb_df.drop(["imdb_title_id"], axis = 1)
encoded_imdb_df.head()
```

```python
encoded_imdb_df.to_csv("Encoded_Data.csv")
```
